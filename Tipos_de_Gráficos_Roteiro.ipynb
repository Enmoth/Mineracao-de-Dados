```python
df = pd.read_csv('vendas_produtos.csv')
df
```

```python
def GraficoLinha(style='dark_background'):
    #estilo de cor
  plt.style.use(style) 

  # Aumentar o tamanho da fonte
  plt.rcParams.update({'font.size': 16})

  # Criar um gráfico de linha
  plt.figure(figsize=(16, 12))

  #descompactar lista
  produto1, = plt.plot(df['Meses'], df['Produto 1'], label='Produto 1', marker='o') 
  produto2, = plt.plot(df['Meses'], df['Produto 2'], label='Produto 2', marker='o')
  produto3, = plt.plot(df['Meses'], df['Produto 3'], label='Produto 3', marker='o')


  plt.legend([produto1, produto2, produto3], ['Produto 1', 'Produto 2', 'Produto 3'])
  plt.title('Vendas de Produtos (2023)')
  plt.xlabel('Mês')
  plt.ylabel('Vendas')

  plt.savefig('GraficoLinha.svg')
  # Mostrar o gráfico
  plt.show()
```

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.normal(0, 100, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

# Interprete a distribuição distribuição e compare a diferença com a distribuição normal

R: distribuição: ela mostra a distribuição de dados da variavel, ela mostra como estão distruibuidos ao longo de diferentes categorias ou intervalos.

normal: média e o desvio padrão determinando a sua localização e dispersão.

```python
# Dados
generos = ['Feminino', 'Masculino']
quantidade = [650, 550]
```

```python
# Criar o gráfico de pizza
fig, ax = plt.subplots(figsize=(8.4, 6.4),facecolor='black')
ax.pie(quantidade, labels=generos, autopct='%1.1f%%', startangle=90, colors=['#FF0000', '#0000FF'], wedgeprops={'width': 0.4})
ax.set_facecolor('black')
ax.set_title('Distribuição de Gênero')

fig.set_facecolor('black')

# Mostrar o gráfico
plt.show()
```

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R:
Com base no gráfico podemos verificar que a discrepancia entre os dados pode estar como valor minimo de aproximadamente -10 e máximo de aproximadamente 90, enquanto temos o primeiro quartil entre 24 e 42 o segundo quartil/mediana sendo um valor aproximado de 43/44 e o terceiro quartil estando aproximadamente entre 44 e 56, onde os quartis estão representando respectivamente 25%, 50% e 75% dos dados da amostra, sendo assim o primeiro quartil demonstra que pelo menos 25% dos dados está abaixo dele, segundo quartil 50% está acima e 50% abaixo enquanto o terceiro quartil trás que 75% dos dados estão abaixo dele, por fim os quartis seriam os dados presentes na "caixa" no entre os pontos máximos e minimos.

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

# Explique o gráfico acima, qual é a diferença da Equipe C da Equipe D.
R:
O gráfico da equipe D trás uma concentração de dados mais precisa, com uma densidade de dados maior devido a seus limite máximos e minímos serem mais próximos.
