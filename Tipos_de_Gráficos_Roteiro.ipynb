```python
def GraficoLinha(style = 'ggplot'):
  #estilo de cor
  plt.style.use(style) 
  # Aumentar o tamanho da fonte
  plt.rcParams.update({'font.size': 16})

  # Criar um gráfico de linha
  plt.figure(figsize=(16, 12))

  #descompactar lista
  produto1, = plt.plot(df['Meses'], df['Produto 1'], label='Produto 1', marker='o') 
  produto2, = plt.plot(df['Meses'], df['Produto 2'], label='Produto 2', marker='o')
  produto3, = plt.plot(df['Meses'], df['Produto 3'], label='Produto 3', marker='o')


  plt.legend([produto1, produto2, produto3], ['Produto 1', 'Produto 2', 'Produto 3'])
  plt.title('Vendas de Produtos (2023)')
  plt.xlabel('Mês')
  plt.ylabel('Vendas')

  plt.savefig('GraficoLinha.svg')
  # Mostrar o gráfico
  plt.show()  
```

# Agora faça o mesmo para o gráfico de barras

```python
import matplotlib.pyplot as plt

def GraficoBarra(style='ggplot'):
    # Estilo de cor
    plt.style.use(style)
    
    # Aumentar o tamanho da fonte
    plt.rcParams.update({'font.size': 16})
    
    # Criar um gráfico de barras
    plt.figure(figsize=(16, 12))
    
    # Descompactar lista
    plt.bar(df['Meses'], df['Produto 1'], label='Produto 1')
    plt.bar(df['Meses'], df['Produto 2'], label='Produto 2')
    plt.bar(df['Meses'], df['Produto 3'], label='Produto 3')
    
    plt.legend()
    plt.title('Vendas de Produtos (2023)')
    plt.xlabel('Mês')
    plt.ylabel('Vendas')
    
    plt.savefig('GraficoBarra.svg')
    # Mostrar o gráfico
    plt.show()
```

```python
GraficoBarra(style='ggplot')
```

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.randint(0, 100, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar um conjunto de dados aleatórios com distribuição normal
np.random.seed(0)
idades = np.random.normal(loc=50, scale=10, size=1000)  # Distribuição normal com média 50 e desvio padrão 10

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

# Explique o gráfico acima, qual é a diferença da Equipe C da Equipe D.
R:

O gráfico gerado é um gráfico de violino que representa a distribuição da pontuação por equipe. Vamos analisar as características e a diferença entre as equipes C e D:

gráfico representa a distribuição da pontuação de uma equipe.
A largura do violino indica a densidade da distribuição de pontuações para cada equipe. Quanto mais largo o violino, maior é a frequência de pontuações em uma determinada faixa.
A linha branca no centro de cada violino representa a mediana da distribuição da pontuação para cada equipe.
As extremidades dos violinos são chamadas de "arestas". Elas mostram a variação das pontuações em cada equipe. Quanto mais longas as arestas, maior é a dispersão dos valores.
Os pontos individuais acima e abaixo dos violinos representam valores discrepantes (outliers) que estão fora do intervalo da maioria das pontuações.
